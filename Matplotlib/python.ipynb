# Matplotlib

- Useful for making Plots, Graphs etc

```python
# importing Matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100) 
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

# Ploting the data

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

### Adding a title, x axis and y axia lables

```python
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine value')
plt.title('Sine Wave')
plt.show()
```

# Ploting a Parabola

```python
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
#ploting with other symbols
plt.plot(x,y, 'r+')
plt.show()
```

```python
plt.plot(x,y, 'g.')
plt.show()
```

```python
plt.plot(x,y, 'rx')
plt.show()
```

## Ploting multiple lines in a single graph

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

# Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]

ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

# PieChart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])

languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]

ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

## Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='g')
ax.scatter(x,z, color='b')
plt.show()           
```

## 3D Scattered plot

```python
fig2 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, c=z, cmap='Blues')
plt.show()
```
