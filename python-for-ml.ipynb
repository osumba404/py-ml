## Constants & Variables

```python
superhero = "Captain America"
print(superhero)
```

```python
superhero = "Batman"
print(superhero)
```

```python
hero1, hero2, hero3 = "Spider Man", "The Joker", "Wonder Woman"
print(hero1)
print(hero2)
print(hero3)
```

```python
x = y = z = 34
print(x)
print(y)
print(z)
```

## Input Function

```python
number1 = int(input("Enter the first no: "))
number2 = int(input("Enter the second no: "))
sum = number1+number2
print(number1)
print(number2)
print("The sum is: " + str(sum))
```
