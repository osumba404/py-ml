# Functions

```python
# a block of code that can be reused in a program

# Factorial of a number - the product of all the positive integers less than or equall to the number

number = int(input("Enter a random number to find its factorial:"))
factorial = 1

if number == 0:
    print("The factorial of 0 is 1")
else:
    for i in range(1, number+1):
        factorial = factorial*i

    print("The factorial of",number," is", factorial)
```

## Factorial function

```python
def factorial_value(num):
    factorial = 1

    if num == 0:
        return factorial
    else:
        for i in range(1, num+1):
            factorial = factorial*i
        return factorial
    
```

```python
print(factorial_value(5))
```
