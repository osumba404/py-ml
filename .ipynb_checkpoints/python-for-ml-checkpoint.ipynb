## Special Data Types in Python

### List

```python
# list should be included in square brackets []
my_list = [1,2,3,4,5]
print(my_list)
type(my_list)
```

```python
#List can have multiple data types
mylist = [2,3,4.9, "Evans", True]
print(mylist)
type(mylist)
```

```python
#List are mutable (are changable)

#adding elements to a list
mylist = [2,3,4.9, "Evans", True]
mylist.append(34)
print(mylist)
print(mylist[4])
print(len(mylist))
```

```python
#initiate an empty list
newlist = []
print(newlist)
```

```python
newlist.append(4)
newlist.append("new word")
print(newlist)
```

```python
#delete an item
mylist = [2,3,4.9, "Evans", True]
del mylist[2]
print(mylist)
```

```python
#join 2 lists
list1 = [2,3,4.9, "Evans", True]
list2 = [2,3,4.9, "Evans", True]

list3 = list1 + list2
print(list3)
```

## Tuple

```python
# Are immutable, cant be changed once created
mytuple = (2,3,4.9, "Evans", True)
print(mytuple)
type(mytuple)
```

### Convert list to tuple

```python
mylist = [2,3,4.9, "Evans", True]
print(mylist)
type(mylist)

mytuple = tuple(mylist)
print(mytuple)
```

## Sets

```python
# also immutable, enclosed in curly brackets {}
#does not allow duplicate/repeated values
myset = {2,3,4.9, "Evans", True}
print(myset)
type(myset)
```

```python
# convert list to set
mylist = [2,3,4.9, "Evans", True]

xhe = set(mylist)
print(xhe)
type(xhe)
```

## Dictionary

```python
# they have a key value pair
#does not allow duplicate/repeated values
mydictionary = {'name':'Evans', 'age':25, 'country':'Kenya'}
print(mydictionary)
type(mydictionary)
```

```python
print(mydictionary['name'])
print(mydictionary['age'])
print(mydictionary['country'])
```
