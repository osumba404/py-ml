# Imbalanced Dataset

```python
# importing libraries/dependancies
import numpy as np
import pandas as pd
```

```python
# loading the dataset to a dataframe
credit_card = pd.read_csv('credit_data.csv')
```

### Separating the legit and fradulent transactions

```python
legit = credit_card[credit_card.Class == 0]
fraud = credit_card[credit_card.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

## Under-sampling

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

```python
legit_sample.shape
```

### Concatenete the two Data Frames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
