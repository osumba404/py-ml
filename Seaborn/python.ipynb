```python
# 1. Define the URL for the original source
data_url = "http://lib.stat.cmu.edu/datasets/boston"

# 2. Read the raw data (it has a tricky format with split lines)
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

# 3. Stitch the split lines together using numpy
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

# 4. Define feature names (standard Boston column names)
feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

# 5. Create the final Pandas DataFrame
boston_df = pd.DataFrame(data, columns=feature_names)
boston_df['MEDV'] = target  # Add the target (Price) column

# View the result
print(boston_df.head())
```

```python
boston_df.head()
```

```python
sns.distplot(boston_df['MEDV'])
```
